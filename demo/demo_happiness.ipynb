```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(happy17)
plt.show()
```

```python
happy17.columns
```

```python
sns.jointplot('Economy..GDP.per.Capita.', 'Trust..Government.Corruption.', data=happy17, size=7, color="g")
plt.show()
```

```python
sns.jointplot('Economy..GDP.per.Capita.',  'Health..Life.Expectancy.', data=happy17, size=7, color="b")
plt.show()
```

```python
sns.distplot(happy17['Health..Life.Expectancy.'], bins = 50, kde = True, rug = True, color = 'm')
plt.show()
```

```python
sns.lmplot('Economy..GDP.per.Capita.',  
           'Health..Life.Expectancy.', 
           data=happy17)
plt.show()
```

```python
sns.lmplot('Freedom',  
           'Health..Life.Expectancy.', 
           data=happy17)
plt.show()
```
